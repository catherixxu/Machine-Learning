```python
import os
import datetime
import numpy
import cv2

import tensorflow
from tensorflow import keras
```

```python
images = []

def load_images_from_folder(folder):
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
        if img is not None:
            images.append(img)
    return images

load_images_from_folder('/Users/catherinexu/Documents/QUT/ML/Data/Q4/UTKFace')
```
