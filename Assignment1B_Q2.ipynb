```python
import pandas
import numpy
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.mixture import GaussianMixture
```

```python
#Import ratings dataset
ratings = pandas.read_csv('/Users/amalieoe/Downloads/Data/Q2/
ratings.csv');
#Import movies dataset
movies = pandas.read_csv('/Users/amalieoe/Downloads/Data/Q2/
movies.csv');
# The number of ratings and the total number of movies
print('The dataset contains: ', len(ratings), ' ratings of ',
len(movies), ' movies.')
```

```python
# Function to get the genre ratings (ignoring the 'no genre' adn'IMAX')
def get_genre_ratings(ratings, movies, genres, column_names):
    genre_ratings = pandas.DataFrame()
    for genre in genres:
        genre_movies = movies[movies['genres'].str.contains(genre) ]
        avg_genre_votes_per_user =
ratings[ratings['movieId'].isin(genre_movies['movieId'])].loc[:, ['userId', 'rating']].groupby(['userId'])['rating'].mean().round(2)
        genre_ratings = pandas.concat([genre_ratings, avg_genre_votes_per_user], axis=1)
    genre_ratings.columns = column_names
    return genre_ratings
```

```python
# Calculate the average rating of movies
genre_ratings = get_genre_ratings(ratings, movies, ['Drama', 'Comedy','Action', 'Thriller', 'Adventure', 'Romance', 'Sci-Fi',
'Crime', 'Fantasy', 'Children', 'Mystery', 'Horror', 'Animation', 'War','Musical','Western','Documentary','Film-Noir'],
['Avg_Drama_Rating', 'Avg_Comedy_Rating','Avt_Action_Rating', 'Avg_Thriller_Rating', 'Avg_Adventure_Rating', 'Avg_Romance_Rating','Avg_Sci-Fi_Rating', 'Avg_Crime_Rating', 'Avg_Fantasy_Rating','Avg_Children_Rating', 'Avg_Mystery_Rating', 'Avg_Horror_Rating','Avg_Animation_Rating','Avg_War_Rating','Avg_Musical_Rating','Avg_Western_Rating','Avg_Documentary_Rating','Avg_Film-Noir_Rating'])

#setting the rating of a genre to 0, if the user hasn't rated any
movies of that genre (Avg_genre_rating=NaN)
genre_ratings = genre_ratings.fillna(0)

print(genre_ratings.head(5))

#Counting number of genres
x=['Drama', 'Comedy','Action', 'Thriller', 'Adventure', 'Romance',
'Sci-Fi', 'Crime', 'Fantasy', 'Children', 'Mystery', 'Horror',
'Animation', 'War','Musical','Western','Documentary','Film-Noir']

#Splitting into test and train: test is the users we are going to
gve recommendations
train=genre_ratings.drop([42,314,444])
test =genre_ratings.loc[[42,314,444],]

#Transforming dataset into a list
X_train = train[['Avg_Drama_Rating',
'Avg_Comedy_Rating','Avt_Action_Rating', 'Avg_Thriller_Rating',
'Avg_Adventure_Rating', 'Avg_Romance_Rating', 'Avg_Sci-Fi_Rating',
'Avg_Crime_Rating', 'Avg_Fantasy_Rating', 'Avg_Children_Rating',
'Avg_Mystery_Rating', 'Avg_Horror_Rating', 'Avg_Animation_Rating',
'Avg_War_Rating','Avg_Musical_Rating','Avg_Western_Rating','Avg_Docu
mentary_Rating','Avg_Film-Noir_Rating']].values
X_test = test[['Avg_Drama_Rating',
'Avg_Comedy_Rating','Avt_Action_Rating', 'Avg_Thriller_Rating',
'Avg_Adventure_Rating', 'Avg_Romance_Rating', 'Avg_Sci-Fi_Rating',
'Avg_Crime_Rating', 'Avg_Fantasy_Rating', 'Avg_Children_Rating',
'Avg_Mystery_Rating', 'Avg_Horror_Rating', 'Avg_Animation_Rating',
'Avg_War_Rating','Avg_Musical_Rating','Avg_Western_Rating','Avg_Docu
mentary_Rating','Avg_Film-Noir_Rating']].values

```

```python
#Calculating best k based on BIC
costs = [];
approx_bic = []

for i in range(100):
    kmeans = KMeans(n_clusters=i+1, random_state=4).fit(X_train)
    k = numpy.shape(kmeans.cluster_centers_)[0]*(numpy.shape(kmeans.cluster_centers_)[1] + 1)
    m = len(X_train)
    costs.append(kmeans.inertia_)
    approx_bic.append(m*numpy.log(kmeans.inertia_ / m) + k*numpy.log(m))
    
    fig = plt.figure(figsize=[8, 6])
ax = fig.add_subplot(1, 1, 1)
ax.plot(approx_bic)
ax.set_xlabel('Number of Clusters')
ax.set_ylabel('Approximate BIC');
print(numpy.argmin(approx_bic) + 1)
```

```python
#Caculating the reconstruction error
recon_error = []
for i in range (50):
    kmeans = KMeans(n_clusters=(i+1), random_state=4).fit(X_train)
    recon_error.append(kmeans.inertia_)

fig = plt.figure(figsize=[8, 6])
ax = fig.add_subplot(1, 1, 1)
ax.plot(recon_error)
ax.set_xlabel('Number of Clusters')
ax.set_ylabel('Reconstruction Error');
"""
#Clustering with k=18 (because we have 18 genres):
kmeans = KMeans(n_clusters=18,random_state=4).fit(X_train)
```
